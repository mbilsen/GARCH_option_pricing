```python
import pandas as pd
import numpy as np
from sklearn.metrics import mean_squared_error
from sklearn.metrics import mean_absolute_error
```

```python
my_options = pd.read_csv("../Daten/different_options_GARCH_precise_conmean0605.csv")
```

```python
my_options.drop("Quote date", axis = 1, inplace = True)
my_options.drop("Expiration date", axis = 1, inplace = True)
my_options.drop("r", axis = 1, inplace = True)
my_options.drop("T", axis = 1, inplace = True)
my_options.drop("GARCH price", axis = 1, inplace = True)
```

```python
my_options["S/K"] = my_options["S"] / my_options["K"]
```

```python
my_options.rename(columns = {'C':'Premium'}, inplace = True)
```

```python
my_options["Days to Mature"].unique()
```

```python
t_list = [1,10,30,45,60,100]
```

```python
for t in t_list:
    my_options1 = my_options[my_options["Days to Mature"] == t]
    my_options1.drop("Days to Mature", axis = 1, inplace = True)
    desc1 = my_options1.describe()
    desc1.drop("count", inplace = True)
    #desc1.rename(index={"count": "N"}, inplace = True)
    #desc1.iloc[0:1] = desc1.iloc[0:1].astype(int)
    desc1 = desc1.round(4)
    index = desc1.index
    index.name = f"T = {t}"
    desc1.to_latex(f"../LaTeX Finance/tables/summary/summaryT{t}.tex")
```
